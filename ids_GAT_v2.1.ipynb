```python
import os

import pandas as pd
import numpy as np

import torch

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from torch_geometric.data import Data
```

```python
DATASET_DIR = "/content/drive/MyDrive/Balanced_6Class_400K"

files = {
    "BENIGN.csv":0,
    "DDoS.csv":1,
    "DoS.csv":1,
    "Mirai.csv":1,
    "Recon.csv":1,
    "Spoofing.csv":1
}

dfs = []

for file,label in files.items():

    path = os.path.join(DATASET_DIR,file)

    df = pd.read_csv(path)

    df["label"] = label

    dfs.append(df)

data = pd.concat(
    dfs,
    ignore_index=True
)

print(data.shape)
```

```python
data = data.sort_values(
    "bidirectional_first_seen_ms"
).reset_index(drop=True)

data["flow_id"] = np.arange(len(data))

print(data.shape)
```

```python
DROP_COLS = [

    "id",
    "expiration_id",

    "src_ip",
    "dst_ip",

    "src_mac",
    "dst_mac",

    "src_oui",
    "dst_oui",

    "requested_server_name",

    "client_fingerprint",
    "server_fingerprint",

    "user_agent",
    "content_type",

    "application_name",
    "application_category_name",

    "flow_id",
    "label",

    "bidirectional_first_seen_ms",
    "bidirectional_last_seen_ms",

    "src2dst_first_seen_ms",
    "src2dst_last_seen_ms",

    "dst2src_first_seen_ms",
    "dst2src_last_seen_ms",

    "application_confidence",
    "application_is_guessed"

]

feature_cols = [

    c
    for c in data.columns
    if c not in DROP_COLS
]

print("Features:",len(feature_cols))
```

```python
for col in feature_cols:

    data[col] = pd.to_numeric(
        data[col],
        errors="coerce"
    )

data[feature_cols] = (
    data[feature_cols]
    .fillna(0)
)
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(
    data[feature_cols]
)

print(X.shape)
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(
    data[feature_cols]
)

print(X.shape)
```

```python
all_hosts = pd.concat(
    [
        data["src_ip"],
        data["dst_ip"]
    ]
).unique()

host_to_id = {

    host : idx + len(data)

    for idx,host
    in enumerate(all_hosts)
}

num_flow_nodes = len(data)

num_host_nodes = len(all_hosts)

print("Flow Nodes:",num_flow_nodes)
print("Host Nodes:",num_host_nodes)
```

```python
edges = []

for flow_id,row in data.iterrows():

    src_host = host_to_id[
        row["src_ip"]
    ]

    dst_host = host_to_id[
        row["dst_ip"]
    ]

    edges.append(
        [flow_id,src_host]
    )

    edges.append(
        [src_host,flow_id]
    )

    edges.append(
        [flow_id,dst_host]
    )

    edges.append(
        [dst_host,flow_id]
    )

print("Host-flow edges built")
```

```python
#print(graph.num_edges)
```

```python
src_groups = data.groupby(
    "src_ip"
)["flow_id"].apply(list)

print(
    "Groups:",
    len(src_groups)
)
```

```python
edge_index = torch.tensor(
    edges,
    dtype=torch.long
).t().contiguous()

print(edge_index.shape)
```

```python
flow_features = torch.tensor(
    X,
    dtype=torch.float
)
```

```python
host_features = torch.zeros(

    (
        num_host_nodes,
        flow_features.shape[1]
    ),

    dtype=torch.float
)

host_counts = np.zeros(
    num_host_nodes
)

for flow_id,row in data.iterrows():

    src_idx = (
        host_to_id[
            row["src_ip"]
        ]
        -
        num_flow_nodes
    )

    dst_idx = (
        host_to_id[
            row["dst_ip"]
        ]
        -
        num_flow_nodes
    )

    host_features[src_idx] += flow_features[flow_id]
    host_features[dst_idx] += flow_features[flow_id]

    host_counts[src_idx] += 1
    host_counts[dst_idx] += 1

for i in range(num_host_nodes):

    if host_counts[i] > 0:

        host_features[i] /= (
            host_counts[i]
        )
```

```python
x = torch.cat(
    [
        flow_features,
        host_features
    ],
    dim=0
)

print(x.shape)
```

```python
flow_labels = torch.tensor(
    data["label"].values,
    dtype=torch.long
)

host_labels = torch.full(
    (
        num_host_nodes,
    ),
    -1,
    dtype=torch.long
)

y = torch.cat(
    [
        flow_labels,
        host_labels
    ]
)

print(y.shape)
```

```python
idx = np.arange(
    num_flow_nodes
)

train_idx,test_idx = train_test_split(

    idx,

    test_size=0.2,

    random_state=42,

    stratify=flow_labels.numpy()
)
```

```python
total_nodes = (
    num_flow_nodes
    +
    num_host_nodes
)

train_mask = torch.zeros(
    total_nodes,
    dtype=torch.bool
)

test_mask = torch.zeros(
    total_nodes,
    dtype=torch.bool
)

train_mask[
    train_idx
] = True

test_mask[
    test_idx
] = True
```

```python
graph = Data(

    x=x,

    edge_index=edge_index,

    y=y,

    train_mask=train_mask,

    test_mask=test_mask
)

print(graph)

print(
    "Nodes:",
    graph.num_nodes
)

print(
    "Edges:",
    graph.num_edges
)
```

```python
device = torch.device(
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)

print(device)
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GATv2Conv

class GATModel(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, num_classes, heads=4):
        super().__init__()

        # Layer 1
        self.conv1 = GATv2Conv(in_channels, hidden_channels // heads, heads=heads)
        self.bn1 = torch.nn.BatchNorm1d(hidden_channels)

        # Layer 2
        self.conv2 = GATv2Conv(hidden_channels, hidden_channels // heads, heads=heads)
        self.bn2 = torch.nn.BatchNorm1d(hidden_channels)

        # Layer 3
        self.conv3 = GATv2Conv(hidden_channels, (hidden_channels // 2) // heads, heads=heads)
        self.bn3 = torch.nn.BatchNorm1d(hidden_channels // 2)

        # Final Classifier
        self.classifier = torch.nn.Linear(hidden_channels // 2, num_classes)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = self.bn1(x)
        x = F.relu(x)
        x = F.dropout(x, p=0.4, training=self.training)

        x = self.conv2(x, edge_index)
        x = self.bn2(x)
        x = F.relu(x)
        x = F.dropout(x, p=0.4, training=self.training)

        x = self.conv3(x, edge_index)
        x = self.bn3(x)
        x = F.relu(x)

        x = self.classifier(x)
        return x
```

```python


model = GATModel(
    in_channels=graph.num_node_features,
    hidden_channels=128,  # Kept exactly as it was in your GraphSAGE notebook
    num_classes=2,
    heads=4
)

model = model.to(device)
graph = graph.to(device)

print(model)
```

```python
optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.002,
    weight_decay=1e-4
)

scheduler = torch.optim.lr_scheduler.StepLR(
    optimizer,
    step_size=20,
    gamma=0.5
)
```

```python
weights = torch.tensor(
    [1.0, 1.5],
    dtype=torch.float
).to(device)

criterion = torch.nn.CrossEntropyLoss(
    weight=weights
)
```

```python
def train():

    model.train()

    optimizer.zero_grad()

    out = model(
        graph.x,
        graph.edge_index
    )

    loss = criterion(
        out[graph.train_mask],
        graph.y[graph.train_mask]
    )

    loss.backward()

    optimizer.step()

    return loss.item()
```

```python
@torch.no_grad()

def evaluate(mask):

    model.eval()

    out = model(
        graph.x,
        graph.edge_index
    )

    pred = out.argmax(
        dim=1
    )

    correct = (
        pred[mask]
        ==
        graph.y[mask]
    ).sum()

    acc = (
        correct.item()
        /
        mask.sum().item()
    )

    return acc
```

```python
for epoch in range(
    1,
    101
):

    loss = train()

    train_acc = evaluate(
        graph.train_mask
    )

    test_acc = evaluate(
        graph.test_mask
    )

    scheduler.step()

    print(
        f"Epoch {epoch:02d}"
        f" | Loss {loss:.4f}"
        f" | Train {train_acc:.4f}"
        f" | Test {test_acc:.4f}"
    )
```

```python
model.eval()

with torch.no_grad():

    out = model(
        graph.x,
        graph.edge_index
    )

    pred = out.argmax(
        dim=1
    )
```

```python
from sklearn.metrics import (
    classification_report
)

print(

    classification_report(

        graph.y[
            graph.test_mask
        ].cpu(),

        pred[
            graph.test_mask
        ].cpu()
    )
)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# True labels
y_true = graph.y[graph.test_mask].cpu().numpy()

# Predicted labels
y_pred = pred[graph.test_mask].cpu().numpy()

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Normalize row-wise
cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

plt.figure(figsize=(7,6))

sns.heatmap(
    cm_percent,
    annot=True,
    fmt=".2%",
    cmap="Greens",
    xticklabels=["Benign", "Attack"],
    yticklabels=["Benign", "Attack"]
)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Normalized Confusion Matrix")

plt.tight_layout()
plt.show()
```
